### < Data Cleaning >

```python
# titanic.csv
import pandas as pd
titanic_df = pd.read_csv("titanic.csv")
```

```python
# info() : Provides essential details about your dataset
titanic_df.info()
```

```python
# drop() : Used to delete columns and rows
titanic_df.drop(columns='Name', inplace=True)
titanic_df.head(5)
```

```python
# Task : Remove ‘Cabin’, ‘Embarked’ and ‘Ticket’ columns
titanic_df.drop(columns=['Cabin', 'Ticket', 'Embarked'], inplace=True)
titanic_df.head(5)
```

#### Handling Duplicate Data 

```python
# 891~894 중복
titanic_df.tail(10)
```

```python
# duplicated() : Used to indicate duplicate values
titanic_df[titanic_df.duplicated()]
```

```python
# drop_duplicates() : Used to remove duplicate rows
titanic_df.drop_duplicates(inplace = True)
titanic_df[titanic_df.duplicated()]
```

```python
titanic_df.tail(10)
```

#### Dealing with Missing Values

```python
# Missing values or null values
# isnull() function can be used to locate the missing values
titanic_df.isnull().head(10)
```

```python
# Summary of missing values
titanic_df.isnull().sum()
```

#### Data imputation

```python
# Imputation : Values are filled in the place of missing data
# Way 1. One popular naïve data imputation is mean substitution : fillna() function to replace null values
titanic_df['Age'].fillna(titanic_df['Age'].mean(), inplace=True)
titanic_df.isnull().sum()
```

```python
# Way1 결과 : index=5의 Age 값이 NaN -> 29.699118 로 변경
titanic_df.head(10)  
```

```python
# Way2. Linear interpolation : Simply means to estimate a missing value by connecting dots in a straight line in increasing order
titanic_df["Age"].interpolate(method = "linear", inplace = True)
```

```python
# Way2 결과 : index=5의 Age 값이 NaN -> 44.5 로 변경
titanic_df.head(10)
```

#### Data omission/deletion


```python
# dropna() function to remove rows that contain null values
titanic_df.dropna(inplace=True)
titanic_df.isnull().sum() # Fare의 11개의 행 제거 
```

### < Data Transformation >

#### Renaming column names


```python
titanic_df.columns
```

```python
# rename() : You can change all column names using dict
titanic_df.rename(columns={'SibSp': 'Sibling.Spouse'}, inplace=True)
titanic_df.columns
```

#### Data encoding

```python
# replace() : Used to replace row values
titanic_df.replace({'Gender': {'male': 0}}, inplace = True)
titanic_df.replace({'Gender': {'female': 1}}, inplace = True)
titanic_df.head(10)
```

### < Data Normalization >

#### Outlier detection

```python
import seaborn as sns # Matplotlib 기반의 시각화 라이브러리
sns.boxplot(y = titanic_df['Fare'] );
```

```python
titanic_df.drop(titanic_df[titanic_df.Fare >= 500].index, inplace = True)
sns.boxplot(y = titanic_df['Fare'] );
```

#### normalization

```python
titanic_df.describe()
```

```python
# Min-max normalization : Using MinMaxScaler() function of sklearn library
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
titanic_df[['Fare']] = scaler.fit_transform(titanic_df[['Fare']])
titanic_df.describe()
```

#### Correlation

```python
sns.heatmap(titanic_df.corr(), annot=True)
```
